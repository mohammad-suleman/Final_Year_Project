```python
from ultralytics import YOLO
import cv2

# Load your trained YOLOv5 model
model = YOLO("/content/drive/MyDrive/FYP-II/best.pt")

# Open the video file
video_path = "/content/drive/MyDrive/FYP-II/test1.mp4"
cap = cv2.VideoCapture(video_path)

# Get original video details
orig_fps = int(cap.get(cv2.CAP_PROP_FPS))
orig_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
orig_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Manually set FPS and resolution
manual_fps = 30  # Change this to your desired FPS
manual_width = 512  # Change to your desired width
manual_height = 512  # Change to your desired height

# Define the output video file (ensure filename is included)
output_path = "/content/drive/MyDrive/FYP-II/output/output_video.mp4"
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter(output_path, fourcc, manual_fps, (manual_width, manual_height))

frame_count = 0  # Counter for processed frames

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Resize the frame to the chosen resolution
    frame = cv2.resize(frame, (manual_width, manual_height))

    # Run YOLOv5 inference on the frame
    results = model(frame)

    # Plot the detections on the frame
    annotated_frame = results[0].plot()

    # Write the frame to the output video
    out.write(annotated_frame)

    frame_count += 1  # Increment frame count

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()

print(f" Processed video saved as: {output_path}")
print(f" Total frames processed: {frame_count}")
print(f" Resolution: {manual_width}x{manual_height} | FPS: {manual_fps}")
```

```python
import cv2
import firebase_admin
from firebase_admin import credentials, firestore
import os
from ultralytics import YOLO
import random

# Correct Firebase JSON path
firebase_json_path = "/mnt/f/final2/testing/roadradar-55dee-firebase-adminsdk-fbsvc-04aa1a4c7e.json"
if not os.path.exists(firebase_json_path):
    raise FileNotFoundError(f"Firebase JSON file not found at: {firebase_json_path}")

# Initialize Firebase
if not firebase_admin._apps:
    cred = credentials.Certificate(firebase_json_path)
    firebase_admin.initialize_app(cred)
db = firestore.client()

# Function to generate random GPS coordinates in Karachi, Pakistan
def get_random_karachi_coordinates():
    lat = round(random.uniform(24.8500, 24.9500), 6)  # Latitude range for Karachi
    lon = round(random.uniform(67.0000, 67.2000), 6)  # Longitude range for Karachi
    return lat, lon

# Correct YOLO Model Path
model_path = "/mnt/f/final2/best_model/yolov5n_finetuned/weights/best.pt"
if not os.path.exists(model_path):
    raise FileNotFoundError(f"YOLO model file not found at: {model_path}")

# Load YOLOv5 model
model = YOLO(model_path)

# Open the video file
video_path = "/mnt/f/final2/testing/test1.mp4"
if not os.path.exists(video_path):
    raise FileNotFoundError(f"Video file not found at: {video_path}")

cap = cv2.VideoCapture(video_path)

# Get video properties
fps = 25  # Set FPS to 25
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define output paths
video_name = os.path.basename(video_path).split('.')[0]  # Extract video name (without extension)
firebase_collection_name = f"Detections_{video_name}"  # Unique collection for each video

output_video_path = f"/mnt/f/final2/testing/output/{video_name}_out.mp4"

# Set up video writer
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

frame_count = 0
batch_logs = []  # Store logs in batches

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO detection
    results = model(frame)

    # Get dummy GPS coordinates
    latitude, longitude = get_random_karachi_coordinates()

    # Process detections
    for result in results:
        for box in result.boxes:
            crack_type = result.names[int(box.cls[0])]  # Get crack type
            confidence = float(box.conf[0]) * 100  # Convert to percentage

            # Prepare log entry
            log_entry = {
                "Frame": frame_count,
                "Crack Type": crack_type,
                "Latitude": latitude,
                "Longitude": longitude,
                "Confidence": confidence
            }

            # Add log entry to batch
            batch_logs.append(log_entry)

    # Annotate frame
    annotated_frame = results[0].plot()

    # Write to video
    out.write(annotated_frame)

    # Upload logs to Firebase in batches (every 10 logs)
    if len(batch_logs) >= 10:
        for log in batch_logs:
            db.collection(firebase_collection_name).add(log)
        batch_logs = []  # Clear batch after upload

    frame_count += 1

# Upload remaining logs (if any) at the end
if batch_logs:
    for log in batch_logs:
        db.collection(firebase_collection_name).add(log)

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()

print(f"Processed video saved at: {output_video_path}")
print(f"Detection logs saved to Firebase collection: {firebase_collection_name}")
```
